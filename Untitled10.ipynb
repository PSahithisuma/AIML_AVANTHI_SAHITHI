step 5:Identify Categorical data and Encode it

```python
dummies = pd.get_dummies(df['State'])
dummies
```

```python
merged=pd.concat([X,dummies],axis='columns')
merged
```

```python
final=merged.drop(['State','California'],axis='columns')
final
```

step 6:Split the Data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(final,y,test_size=0.2)
```
