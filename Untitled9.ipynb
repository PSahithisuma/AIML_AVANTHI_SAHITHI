```python
import numpy as np
import matplotlib.pyplot as plot

x = np.arange(0,5,0.1)
y =np.sin(x)
plt.plot(x,y)

plt.title('Our first Sine Graph',fontdict={'fontsize':100})
```

```python
x=[1,2,3]
y=[4,5,6]

plt.plot(x,y)

plt.title('Our First Sine Graph',fontdict={'fontsize':10})

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
x2 = np.arange(0,4.5,0.5)
plt.plot(x2,x2**2, 'y', label='x^2')

plt.legend()
```
