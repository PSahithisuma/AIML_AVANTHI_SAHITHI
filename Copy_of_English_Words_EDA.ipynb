Import the Pandas library

```python
import pandas as pd
```

Read the txt file into a dataframe

```python
 df = pd.read_csv('/content/unix-words.txt')
 df
```

Display the first 10 rows of the dataframe

```python
df.head(10)
```

Change the column name 'A' to the words

```python
df.rename(columns={'A':'Word'},inplace = True )
df
```

Add character count of each word as another column in the dataframe

```python
df['Word']=df['Word'].fillna('')
df['CharCount'] = df['Word'].apply(len)
df
```

Add a column called value which is the sum of the numerical positions of the alphabets in the word
(Ex: aah = 1 + 1 + 8 = 10)

```python
def summing(Word):
  sum = 0
  Word.lower()
  for char in Word:
    s= ord(char)-96
    sum+=s
  return sum

df['values']=df['Word'].apply(summing)
```

Write a command to find the number of items in the dataframe

```python
df.info
```

Set the words as the index of the dataframe

```python
df.set_index('Word')
df
```
