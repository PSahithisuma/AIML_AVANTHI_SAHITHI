```python
import cv2
```

```python
image = cv2.imread('/content/Carimg.jpg')
p = cv2_imshow(image)
p
```

```python

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
c = cv2_imshow( gray_image)
c
```

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
rgb_image
```

```python
bfilter = cv2.bilateralFilter(gray_image,15,75,75)
bfilter
```

```python
threshold1 = 150
threshold2 = 50
edged = cv2.Canny(bfilter, threshold1, threshold2)
e = cv2_imshow(edged)
e
```
